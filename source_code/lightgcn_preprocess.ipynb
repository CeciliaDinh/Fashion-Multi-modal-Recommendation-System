```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
```

```python
df= pd.read_csv('clean_data.csv')
```

```python
df_lightgcn= df[['user_id','asin','rating','review_date']].copy()
```

```python
# rename variables 
df_lightgcn.rename(columns={'asin':'item_id',
                   'review_date':'timestamp'}, inplace=True)
```

```python
df_lightgcn['timestamp']= pd.to_datetime(df_lightgcn['timestamp'], errors='coerce')
df_lightgcn = df_lightgcn.dropna(subset=['timestamp'])
df_lightgcn['timestamp'] = df_lightgcn['timestamp'].astype('int64') // 10**9
```

```python
user_encoder = LabelEncoder()
item_encoder = LabelEncoder()
```

```python
df_lightgcn['user_id'] = user_encoder.fit_transform(df_lightgcn['user_id'])
df_lightgcn['item_id'] = item_encoder.fit_transform(df_lightgcn['item_id'])
```

```python
df_lightgcn = df_lightgcn.sort_values(by='timestamp')
```

```python
output_path = 'interactions.csv'
df_lightgcn.to_csv(output_path, index=False)
```
